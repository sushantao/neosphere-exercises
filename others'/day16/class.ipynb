```
class Car:
  def __init__(self, name, model=2018):
    '''
      This is a Car class representing real world object
    '''
    self.name = name
    self.model = model
    self.status = 'Stop'
    self.type= None
    
  def start(self):
    '''
      Change status to start
    '''
    self.status = 'Start'
    
  def stop(self):
    '''
      Change status to stop
    '''
    self.status = 'Stop'
    
c = Car('Honda')
print(c.name)
print(c.model)
print(c.__dict__)
c.start()
print(c.__dict__)
c.stop()
print(c.__dict__)
```
